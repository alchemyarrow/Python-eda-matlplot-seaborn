import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```python
df=pd.read_csv('insurance.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df_setosa=df.loc[df['smoker']=='no']
df_virginica=df.loc[df['smoker']=='yup']
df_versicolor=df.loc[df['smoker']=='no']
```

```python
plt.plot(df_setosa['age'],np.zeros_like(df_setosa['age']),'o')
plt.plot(df_virginica['age'],np.zeros_like(df_virginica['age']),'o')
plt.plot(df_versicolor['age'],np.zeros_like(df_versicolor['age']),'o')
plt.xlabel('age')
plt.show()
```

```python
sns.FacetGrid(df,hue="sex",size=5).map(plt.scatter,"charges","region").add_legend();
plt.show()
```

```python
sns.pairplot(df,hue="age",size=3)
```
